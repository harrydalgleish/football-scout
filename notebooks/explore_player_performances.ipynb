```python
import json
import pandas as pd

with open(f"/workspaces/football-scout/data/raw/events/{match_id}.json", "r") as file:
    result = json.load(file)

print(type(result))
print(len(result))
print(result[0])
```

```python
#Code to test all events available in the code
list_of_all_events = []
for event in result:
    event_type = event["type"]["name"]
    if event_type not in list_of_all_events:
        list_of_all_events.append(event_type)
    if event_type == "Pass":
        print(event["pass"])
print(list_of_all_events)
```
